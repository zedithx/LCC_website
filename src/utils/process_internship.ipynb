```python
import pandas as pd
```

```python
df_spring = pd.read_excel('Copy of LCC Spring 2023.xlsx')
```

```python
print(df_spring)
```

```python
df_summer = pd.read_excel('lcc SUMMER COPY (WEBSITE).xlsx')
```

```python
def process_df(df):
    json_list = []
    for index, row in df.iterrows():
        json_list.append(
            {
                'title': row['Company'], 
                'number': row['New booth number (website)'],
                'description': row['Industry']
            })
    for json_object in json_list:
        print('{')
        for key, value in json_object.items():
            print(f'\t{key}: "{value}",')
        print('},')
```

```python
process_df(df_spring)
```

```python
process_df(df_summer)
```
